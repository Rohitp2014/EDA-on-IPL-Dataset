# Exploratory Data Analysis on IPL Dataset

```python
#Import necessary python libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import dataset.
ipl = pd.read_csv("E:\Datasets\matches.csv")
```

```python
ipl.head()
```

```python
ipl.shape
```

```python
#Getting the frequency of most man of the match awards.
ipl['player_of_match'].value_counts()
```

```python
#Getting top 10 players with most man of the match awards.
ipl['player_of_match'].value_counts()[0:10]
```

```python
#Getting list of top 5 players who won most man of the match awards.
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
#Bar-plot for top 5 players with most man of the match awards.
plt.figure(figsize=(7,5))
cl=["r","y","m","k","c"]
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color=cl, width=0.5)
plt.xlabel("Players")
plt.ylabel("Number of Awards")
plt.title("Top 5 players with most man of the match awards")
plt.show()
```

```python
#Getting the frequency of result column
ipl['result'].value_counts()
```

```python
#Finding out the number of toss ins w.r.t. each team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the records where team won on batting first.
batting_first=ipl[ipl['win_by_runs']!=0]
batting_first.head(10)
```

```python
#Making a hisogram for showing frequency of matches win by runs.
plt.figure(figsize=(7,6))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of Runs')
plt.ylabel("No. of matches")
plt.xlabel('Runs')
plt.show()
```

```python
#Finding out the number of wins w.r.t. each team after batting first
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize=(6,5))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["skyblue","yellow","red"], width=0.5)
plt.title("Top 3 teams winning with batting first")
plt.ylabel("No. of winning matches")
plt.show()
```

```python
#Making a pie-chart.
plt.figure(figsize=(6,6))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#Extracting those records where team has won after batting second.
batting_second=ipl[ipl['win_by_wickets']!=0]
batting_second.head(10)
```

```python
#Making a histogram for frequency of wins w.r.t. number of wickets.
plt.figure(figsize=(6,5))
plt.hist(batting_second['win_by_wickets'],bins=25)
plt.xlabel("No. of wickets")
plt.ylabel("No. of winning matches")
plt.show()
```

```python
#Finding out the frquency of number of wins w.r.t. each team after batting second.
batting_second['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after batting second.
plt.figure(figsize=(6,6))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["purple","skyblue","red"], width=0.5)
plt.title("Top 3 teams winning with batting second")
plt.ylabel("No. of winning matches")
plt.show()
```

```python
#Making a pie-chart for distribution of most wins after batting second.
plt.figure(figsize=(6,6))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#Find number of matches played by each season.
ipl['season'].value_counts()
```

```python
#Bar-plot for showing number of matches by each season.
# Count the number of matches per season
season_matches = ipl['season'].value_counts().sort_index()  # Sorting by season

plt.figure(figsize=(10, 6))
sns.barplot(x=season_matches.index, y=season_matches.values, palette="viridis")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.title("Number of Matches Played in Each Season")
plt.xticks(rotation=45)
plt.show()
```
